```python
#Question 1.c
u, s, v = np.linalg.svd(im, full_matrices=True)
```

```python
print("Shape of u matrice :" ,u.shape)
print("Shape of s matrice :" ,s.shape)
print("Shape of v matrice :" ,v.shape)
```

```python
#Question 1.d
ranklist = [3, 10, 20, 30, 50, 60 ,70, 80, 100]
n=[]
error=[]
memory=[]
for i in (ranklist):
    reconstimg = np.matrix(u[:, :i]) * np.diag(s[:i]) * np.matrix(v[:i, :])
    entries = (2*i*512 + i) / (512*512)
    mse = np.square(np.subtract(imarray, reconstimg)).mean()
    #mse /= (512*512) for images, divide the error by size of the image. But it will not change the exhibition.
    title = "n = %s" % i ,"Error = %s" % mse, "Size to original data = %s"  %entries
    n.append(i)
    error.append(mse)
    memory.append(entries)
    counter
    print(title)
```
